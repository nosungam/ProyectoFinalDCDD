```python
# ============================
# Carga de la base de datos desde Google Drive
# ============================

from google.colab import drive
import pandas as pd

# Montar Google Drive
drive.mount('/content/drive')

# Ruta al archivo en tu Google Drive
# Cambiá la ruta según la ubicación real de tu CSV
ruta = '/content/drive/MyDrive/Proyecto_final.csv'

# Cargar la base de datos, specifying the separator
df = pd.read_csv(ruta, sep=';')

# Verificar la carga
print("\n✅ Base cargada correctamente")
print("Dimensiones de la base:", df.shape)
print("\nPrimeras filas de la base:")
print(df.head())
```

```python
# ============================
# Exploración de la base de datos
# ============================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Copiamos el dataframe original y eliminamos las columnas indicadas
df_explore = df.drop(columns=["Id", "Incorporacion", "Formulario",
                              "Horadeinicio", "Horadefinalización",
                              "HoraRealizacion", "TiempoRequerido", "CI"],
                     errors="ignore")

print("✅ Columnas eliminadas. Ahora quedan:", df_explore.shape[1], "variables.\n")

# 1. Dimensiones y estructura
print("Dimensiones de la base:", df_explore.shape)
print("\nPrimeras filas:")
display(df_explore.head())

# 2. Tipos de variables
print("\nTipos de datos por columna:")
print(df_explore.dtypes)

# 3. Valores nulos
print("\nCantidad de valores nulos por columna:")
print(df_explore.isnull().sum())

# 4. Duplicados
print("\nCantidad de filas duplicadas:", df_explore.duplicated().sum())

# 5. Estadísticos descriptivos
print("\nEstadísticos descriptivos - Variables numéricas:")
display(df_explore.describe())

print("\nEstadísticos descriptivos - Variables categóricas:")
display(df_explore.describe(include="object"))

# 6. Distribución de variables numéricas
# Check if there are any numeric columns before plotting histograms
num_cols = df_explore.select_dtypes(include=np.number).columns
if len(num_cols) > 0:
    df_explore[num_cols].hist(figsize=(15, 10), bins=30, edgecolor="black")
    plt.suptitle("Distribución de variables numéricas", fontsize=16)
    plt.show()
else:
    print("\nNo numeric columns found for plotting histograms.")


# 7. Distribución de variables categóricas
# Check if there are any object columns before iterating
cat_cols = df_explore.select_dtypes(include="object").columns
if len(cat_cols) > 0:
    for col in cat_cols:
        plt.figure(figsize=(8,4))
        sns.countplot(data=df_explore, x=col, order=df_explore[col].value_counts().index)
        plt.title(f"Distribución de {col}")
        plt.xticks(rotation=45)
        plt.show()
else:
    print("\nNo categorical columns found for plotting.")

# 8. Correlaciones entre variables numéricas
# Check if there are any numeric columns before calculating correlation
if len(df_explore.select_dtypes(include=np.number).columns) > 0:
    plt.figure(figsize=(12,8))
    corr = df_explore.corr(numeric_only=True)
    sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
    plt.title("Matriz de Correlación")
    plt.show()
else:
    print("\nNo numeric columns found for correlation analysis.")
```

```python
# ============================
# Estadísticas descriptivas en cuadros
# ============================

import pandas as pd

# Variables numéricas
num_stats = df_explore.describe().T   # Transpuesta para verlo como cuadro
print("📊 Estadísticas descriptivas - Variables numéricas")
display(num_stats)

# Variables categóricas
cat_stats = df_explore.describe(include="object").T
print("\n📊 Estadísticas descriptivas - Variables categóricas")
display(cat_stats)
```

```python
# ============================
# Estadísticas de dispersión y asimetría
# ============================

import pandas as pd

# Seleccionamos solo variables numéricas
num_cols = df_explore.select_dtypes(include="number")

# Calculamos medidas adicionales
dispersion_stats = pd.DataFrame({
    "Varianza": num_cols.var(),
    "Desvío Estándar": num_cols.std(),
    "Skewness (Asimetría)": num_cols.skew(),
    "Curtosis": num_cols.kurt()
})

print("📊 Estadísticas de dispersión y asimetría - Variables numéricas")
display(dispersion_stats)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Lista de variables de interés
vars_mbi = ["MBI_1","MBI_2","MBI_3","MBI_4","MBI_5","MBI_6","MBI_7","MBI_8","MBI_9",
            "MBI_10","MBI_11","MBI_12","MBI_13","MBI_14","MBI_15","MBI_16","MBI_17",
            "MBI_18","MBI_19","MBI_20","MBI_21","MBI_22","AE_MBI","DSP_MBI","RP_MBI",
            "Nivel_AE_MBI","Nivel_DSP_MBI","Nivel_RP_MBI","MBI_Burnout","MBI_RiesgoBurnout1"]

# Seleccionamos solo estas columnas
df_mbi = df_explore[vars_mbi].copy()

# Reemplazamos valores 999 por NaN
df_mbi.replace(999, np.nan, inplace=True)

# Eliminamos outliers usando IQR
num_cols = df_mbi.select_dtypes(include=np.number).columns
for col in num_cols:
    Q1 = df_mbi[col].quantile(0.25)
    Q3 = df_mbi[col].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    df_mbi = df_mbi[(df_mbi[col] >= limite_inferior) & (df_mbi[col] <= limite_superior)]

# Calculamos matriz de correlación
corr_matrix = df_mbi.corr()

# Creamos máscara para mostrar solo la mitad superior
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Mapa de calor
plt.figure(figsize=(14,12)) # Aumentamos el tamaño del gráfico
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0, mask=mask, cbar=True,
            annot_kws={"size": 8}) # Disminuimos el tamaño de la fuente de las anotaciones
plt.title("Mapa de Calor de Correlación - Variables MBI", fontsize=16)
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
